```
#import the library  for the scrapying the website 

import requests
```

```
#link of the website 

wiki_link ='https://en.wikipedia.org/wiki/List_of_Asian_countries_by_area'
link=requests.get(wiki_link).text
```

```
print(link)
```

```
from bs4 import BeautifulSoup
soup=BeautifulSoup(link,'lxml')
print(soup)
```

```
print(soup.prettify())
```

```
soup.title
```

```
#what is i dont want the these html tag 

soup.title.string
```

```
#lets found that how many link in this webpage 

soup.a
```

```
soup.find_all('a')
```

```
all_link=soup.find_all('a')
for link in all_link:
  print(link.get("href"))
```

```
all_table=soup.find_all('table')
print(all_table)
```

```
right_table= soup.find('table',class_="wikitable sortable") 
right_table
```

```
table_link=right_table.findAll('a')
table_link
```

```
country=[]
for links in table_link:
  country.append(links.get('title'))
print(country)
```

```
import pandas as pd
df=pd.DataFrame()
df['country']=country
df
```
