```
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```
# y = 3x1 + 5x2
x1 = 10 
x2 = 5 
y_a  = 55
```

```
np.random.randint(0, 10, 1)[0]  # low, high, size
```

```
w1 = np.random.randint(1, 50, 1)[0]
w2 = np.random.randint(1, 50, 1)[0]
print(f"Random Value of w1 = {w1} & w2 = {w2}")
epoch = 30
lr = 0.001 # Generic LR but can be changed for different use cases
error_list = []

for i in range(epoch):
  y_p = w1*x1 + w2*x2 
  print(f"Value of w1 = {w1} and w2 = {w2} and y_p = {y_p}")
  
  # Calculate Error
  error = (y_a - y_p)**2         # yp == ya
  error_list.append(error) 
  print(f"Error = {error}")
  dEdw1 = 2*(y_a - y_p)*(-x1)
  dEdw2 = 2*(y_a - y_p)*(-x2)

  w1 = w1 - lr*dEdw1
  w2 = w2 - lr*dEdw2
```

```
plt.plot(error_list)
plt.xlabel("Epochs")
plt.ylabel("Error")
plt.title("Error Per Epoch")
plt.show()
```
